```python
##2.Is there any correlation between Geographic location and Hospital performance for CABG Surgeries?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('California Hospitals.csv',encoding='latin1')

plt.figure(figsize=(10,6))
sns.scatterplot(x='Longitude',y='Latitude',hue='Performance Rating',data=df)
plt.title('Geographic Location vs. Hospital Performance for CABG Surgeries')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Performance Rating')
plt.show()
```

```python
## 3.What is the state-wide variation in the rate of adverse events following coronary artery bypass grafting (CABG) surgeries between the various counties and hospitals?


df['Hospital Observed Adverse Events Rate (%)']=pd.to_numeric(df['Hospital Observed Adverse Events Rate (%)'],errors='coerce')

df=df.dropna(subset=['Hospital Observed Adverse Events Rate (%)'])

plt.figure(figsize=(12,6))
sns.boxplot(x='County',y='Hospital Observed Adverse Events Rate (%)',data=df,order=df.groupby('County')['Hospital Observed Adverse Events Rate (%)'].median().sort_values().index)
plt.title('State-wide Variation in Adverse Events Rate by County')
plt.xlabel('County')
plt.ylabel('Observed Adverse Events Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
##1.What is the Correlation between different performance metrics at the hospitals?
plt.figure(figsize=(12, 6))
sns.countplot(x='County', hue='Performance Rating', data=df)
plt.title('Distribution of Performance Ratings per County')
plt.xlabel('County')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Performance Rating')
plt.show()
```
